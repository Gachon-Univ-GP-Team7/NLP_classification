```python
import numpy as np
import nltk
import pandas as pd
from konlpy.tag import Okt
from konlpy.tag import Kkma
##코드 최적화는 나중에 진행 예정


df = pd.read_excel('C:/Users/user/Desktop/nlp/data.xlsx')
# print(df)
#나이와 문장만 추출한 데이터 프레임
data = df.loc[:,['STT', 'AGE']]
print(data)

#각 나이별 문장 추출
thirdData = df[df.AGE == 3]
thirdData = thirdData['STT']
print(thirdData)

okt = Okt()
kkma = Kkma()
thirdMorphs=[]      #문장의 형태소를 담은 변수
thirdNouns=[]       #문장의 명사를 담은 변수
thirdNounCnt=[]
thirdMorphsCount=[] #문장내, 형태소의 숫자를 담은 변수
maxMorphCnt = 0 #문장내, 최대 형태소
minMorphCnt = 10 #문장내, 최소 형태소
total = 0

#형태소 평균, 최장, 최단 길이 추출
#낱말 평균, 최장, 최단 길이 추출

##3살 분류##
for i in range(len(thirdData)):
    m = okt.morphs(thirdData[i])
    thirdNouns = okt.nouns(thirdData[i])
    thirdMorphs.append(m)
    thirdMorphsCount.append(len(m))

print("품사 묶음:",thirdMorphs)
print("품사의 수:", thirdMorphsCount)

for j in range(len(thirdMorphsCount)) :
    if maxMorphCnt < thirdMorphsCount[j]:
        maxMorphCnt = thirdMorphsCount[j]
    if minMorphCnt > thirdMorphsCount[j]:
        minMorphCnt = thirdMorphsCount[j]
    total = total + thirdMorphsCount[j]
avg = total/len(thirdMorphsCount)
print("3살 평균 형태소 길이 : ", avg)
print("3살 최대 형태소 길이:", maxMorphCnt)
print("3살 최소 형태소 길이:", minMorphCnt)

fifthData = df[df.AGE == 5]
fifthData = fifthData['STT']
sixthData = df[df.AGE == 6]
sixthData = sixthData['STT']




```
